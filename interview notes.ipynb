```python
def flatten_dict(d, parent_key='', sep='_'):
    items = []
    for k, v in d.items():
        new_key = parent_key + sep + k if parent_key else k
        if isinstance(v, dict):
            items.extend(flatten_dict(v, new_key, sep=sep).items())
        else:
            items.append((new_key, v))
    return dict(items)
```

```python

# Example Usage:
nested_dict = {
    'a': 1,
    'b': {
        'c': 2,
        'd': {
            'e': 3,
            'f': 4
        }
    },
    'g': 5
}

flattened = flatten_dict(nested_dict)
print(flattened)
```

```python
inp = [1,1,1,2,2,3,3,4,4,1]   # expected output =[[1,1,1],[2,2],[3,3],[4,4],[1]]
def split_consecutive(inp_list):
    if not inp_list:
        return []

    result = []
    current_sublist = [inp_list[0]]

    for i in range(1, len(inp_list)):
        if inp_list[i] == inp_list[i-1]:
            current_sublist.append(inp_list[i])
        else:
            result.append(current_sublist)
            current_sublist = [inp_list[i]]
            
    # Append the last sublist after the loop finishes
    result.append(current_sublist) 
    
    return result

print(split_consecutive(inp))
```

```python
#Write a generator to produce Fibonacci numbers

#n = int(input('Give amount: '))

def fib():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

f = fib()

for i in range(10):
    print( next(f))
```

```python
#write a python code to find the first non-repeating character in a string
def fnrepchar(str):
    for i in range(len(str)):
        found = False
        for j in range(len(str)):
            if i != j and str[i] == str[j]:
                found = True
                break
        if not found:
            return str[i]
    return '$'
print(fnrepchar('aabbaccedd'))
```

```python
# FIBONACCI SEQUENCE ITERATIVE APPROACH
def fibonacci_series(n):
    """Prints the first n numbers of the Fibonacci series."""
    a, b = 0, 1
    for _ in range(n):
        print(a, end=' ')
        a, b = b, a + b

# Example usage:
fibonacci_series(10)
```

```python
# FIBONACCI SEQUENCE RECURSIVE APPROACH
def fibonacci_recursive(n):
    """Returns the nth Fibonacci number."""
    if n <= 1:
        return n
    else:
        return fibonacci_recursive(n-1) + fibonacci_recursive(n-2)

# Example to print first 10 terms:
for i in range(10):
    print(fibonacci_recursive(i), end=' ')
```

```python
# Fibonaaci using Dynamic programming

FibArray = [0, 1]
def fibonacci(n):
    if n<= len(FibArray):
        return FibArray[n-1]
    else:
        temp_fib = fibonacci(n-1)+fibonacci(n-2)
        FibArray.append(temp_fib)
        return temp_fib
fibonacci(6)
```
