```python
from collections import defaultdict

words = ["apple", "bat", "car", "banana", "cup"]
res = defaultdict(list)

for word in words:
    res[len(word)].append(word)

print(dict(res))
```

```python
words = ["apple", "bat", "car", "banana", "cup"]
res = {}

for word in words:
    res.setdefault(len(word), []).append(word)

print(res)
```

```python
words = ["apple", "bat", "car", "banana", "cup"]
res = {len(w): [x for x in words if len(x) == len(w)] for w in words}

print(res)
```

```python
# to return top 2 elements of dictionary based on value

inp = ['c','b','a','b','c','a','b', 'a', 'a']
count = {}
print(count)
for w in inp:
    count[w] = 1+ count.get(w,0)
print(count)
res = []
for i in range(2):
    for key, value in count.items():
        if value == max(count.values()):
            res.append(key)
            print(res)
    del count[key]
print(res)
```

```python
# to return top 2 elements of dictionary based on value

inp = ['c','b','a','b','c','a','b', 'a', 'a']
count = {}
print(count)
for w in inp:
    count[w] = 1+ count.get(w,0)
print(count)
res = []
for i in range(0,2):
    res.append([key for key,value in count.items() if value == max(count.values())])
    #count.pop(key, None)
print(res)
```

```python
from collections import Counter
inp = ['c','b','a','b','c','a','b', 'a', 'a']
dic_op = Counter(inp)
print(dict(dic_op))
arr = Counter(inp).most_common(2) # Returns [('a', 4), ('b', 3)]
print(arr)
res = []
for i in range(2):
    res.append(arr.pop()[0])  
print(res)
```
